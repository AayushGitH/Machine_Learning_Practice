```python
import matplotlib.pyplot as plt
# Writing this line so we don't have to write .show() everytime to show graph
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('FirstGraph.png')
```

```python
plt.plot(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y=x*x
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
# Creating subplots
# First subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r*')

# Second subplot
plt.subplot(2,2,2)
plt.plot(x,y,'b')

# Third subplot
plt.subplot(2,2,3)
plt.plot(x,y,'g--')
```

```python
# Creating sine wave form plot
x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form plot')
plt.plot(x,y,'r')
```

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,4,11]
y2 = [6,15,7]
plt.bar(x,y,color='r')
plt.bar(x2,y2,color='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar graph')
```

```python
# Creating histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,color='black')
plt.title('Histogram')
plt.show()
```

```python
# Creating boxplot

data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

```python
# Creating Pie chart

labels = 'Python','C#','Ruby','Java'
sizes = [215,500,245,124]
colors = ['gold', 'yellowgreen', 'lightcoral','lightskyblue']
explode = (0.1,0.1,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()

```
